1. Find all digits in text

```python
import re

text = "My Mobile Number is 805502218080550221180"
#numbers = re.findall(r"\d+", text)
numbers = re.findall(r"\d{10}", text)
print("Digits found:", numbers)
```

2. Find all words starting with a capital letter

```python
import re

text = "Riya and Rahul are in Fergusson College"
result = re.findall(r"\b[A-Z]\w+", text)
print("Words starting with capital letters:", result)
```

3. Extract all vowels

```python
import re

text = "Welcome to Python Programming"
vowels = re.findall(r"[aeiouAEIOU]", text)
print("Vowels:", vowels)
```

4. Check if a string starts with 'Hello'

```python
import re

text = "Hello Students, welcome!"
if re.match(r"^Hello", text):
    print("Yes, string starts with 'Hello'")
else:
    print("No, doesn't start with 'Hello'")
```

OR

```python
import re

text = "Hi Students, welcome!"
if re.match(r"^(Hello|Hi)", text):
    print("Yes, string starts with 'Hello' or 'Hi'")
else:
    print("No, doesn't start with 'Hello'")
```

5. Split a string by commas or semicolons

```python
import re

text = "red;blue,green;yellow"
split_text = re.split(r"[;,]", text)
print("Split result:", split_text)
```

6. Extract all 10-digit phone numbers

```python
import re

text = "Call 9876543210 or 9123456789 today."
phones = re.findall(r"\b\d{10}\b", text)
print("Phone numbers found:", phones)
```

7. Replace all digits with #

```python
import re

text = "There are 3 cats and 12 dogs."
new_text = re.sub(r"\d", "#", text)
print("After replacement:", new_text)
```

8. Validate an email address

```python
import re

email = input("Enter email: ")
pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
```

OR

```python
import re

email = input("Enter email: ")
pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
pattern1=r'^[A-Za-z0-9]+@[A-Za-z0-9]+\.[A-Za-z]+$'

if re.match(pattern1, email):
    print("Valid Email")
else:
    print("Invalid Email")
```

9. Find words starting with ‘a’

```python
import re

text = "apple and Happy yess Apricot are amazing"
words = re.findall(r"\ba\w*", text)
print("Words starting with 'a':", words)
```

10. Extract all dates (DD-MM-YYYY format)

```python
import re

text = "Events on 12-09-2023 and 25-12-2024."
dates = re.findall(r"\b\d{2}-\d{2}-\d{4}\b", text)
print("Dates found:", dates)
```

11. Password strength checker

```python
import re

password = input("Enter password: ")
pattern = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@#$]).{8,}$'

if re.match(pattern, password):
    print("Strong Password")
else:
    print("Weak Password")
```

12. Validate Indian PAN card number

```python
import re

pan = input("Enter PAN number: ")
pattern = r'^[A-Z]{5}\d{4}[A-Z]$'

if re.match(pattern, pan):
    print("Valid PAN Number")
else:
    print("Invalid PAN Number")
```

13. Find consecutive repeated words

```python
import re

text = "This is is a test test sentence"
repeated = re.findall(r"\b(\w+)\s+\1\b", text)
print("Repeated words:", repeated)
```

14. Extract domain names from emails

```python
import re

emails = "abc@gmail.com, xyz@yahoo.in, test@outlook.com"
domains = re.findall(r"@([\w\.]+)", emails)
print("Domains:", domains)
```

15. Find all words ending with 'ing'

```python
import re

text = "He is playing, running and singing."
words = re.findall(r"\w+ing\b", text)
print("Words ending with 'ing':", words)
```

16. Validate IP address (simple version)

```python
import re

ip = "192.168.1.1"
pattern = r"^(\d{1,3}\.){3}\d{1,3}$"
if re.match(pattern, ip):
    print("Valid IP address")
else:
    print("Invalid IP address")
```

17. Extract hashtags from a tweet

```python
import re

tweet = "Loving #Python and #MachineLearning at #FergussonCollege!"
hashtags = re.findall(r"#\w+", tweet)
print("Hashtags:", hashtags)
```

18. Validate URL

```python
import re

url = "https://www.google.com"
pattern = r"^(https?://)?(www\.)?[\w\-]+\.\w+(/[\w\-]*)*/?$"
if re.match(pattern, url):
    print("Valid URL")
else:
    print("Invalid URL")
```
