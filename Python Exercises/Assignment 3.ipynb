```python
# 1.Define a class Box,having members as:
#Data members:length,breadth and height of int type.
#Members functions:
#1.Constructor to initalise the data members.
#2.To compute and display the volume.


class Box:
    def __init__(self,length:int,breadth:int,height:int):
        self.length=length
        self.breadth=breadth
        self.height=height
        
    def disp(self):
        volume=self.length*self.breadth*self.height
        print("Volume will be: ",volume)
        
p1=Box(2,3,4)
p1.disp()
```

```python
#2.Define a class student with rollno,name and marks of 3 subjects.Accept details of student using constructor.
#Write a function to calculate the percentage of a student and display the details of the student.


class student:
    def __init__(self,rollno,name,marks):
        self.rollno=rollno
        self.name=name
        self.marks=marks
        
    def percent(self):
        total_marks=sum(self.marks)
        percentage=total_marks/3
        return percentage
    
    def details(self):
        percentage=self.percent()
        print("Roll no: ",self.rollno)
        print("Name: ",self.name)
        print("Marks: ",self.marks)
        print(f"Percentage: {percentage: .2f}%")

print("\nDetails of student1: \n")        
student1=student(101,"Chetana",[87,97,90])
student1.details()
 
print("\nDetails of student2: \n")  
student2=student(102,"Yami",[90,90,90])
student2.details()

print("\nDetails of student3: \n")  
student3=student(103,"Talat",[99,95,90])
student3.details()
    
```

```python
#3.Write a python program to create a class Bank with data members as account no,customername and balance.
#Define functions like check balance(),deposit(),withdraw().Accept details from the user and define appropriate constructor.

class Bank:
    def __init__(self,account_no,customer_name,balance):
        self.account_no=account_no
        self.customer_name=customer_name
        self.balance=balance
        
    def check_balance(self):
        print("Account Number: ",account_no)
        print("Account Customer: ",customer_name)
        print("Current Balance: ",balance)
        
    def deposit(self,amount):
        if amount>0:
            self.balance += amount
            print("Deposited amount: ",amount)
            print("New balance after adding deposit amount: ",self.balance)
        else:
            print("Deposit amount should be positive only.")
            
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient Balance")
        elif amount > 0:
            self.balance -= amount
            print("Withdraw amount: ",amount)
            print("New balance after withdrawing amount: ",self.balance)
        else:
            print("Withdraw amount should be positive only.")
            
account_no=int(input("Enter the account number: "))
customer_name=input("Enter the customer name: ")
balance=float(input("Enter the balance: "))
print("\n-----Customer Details------\n")
            
obj=Bank(account_no,customer_name,balance)
obj.check_balance()
obj.deposit(1000)
obj.withdraw(500)

            
        
        
```

```python
#4.Define a class mobike with the following description:
#varibales:bn0,phno,name,days,charge
#member methods:void input() and compute().
#write a script to create instance of a mobike class and call above functionality.

class Mobike:
    def __init__(self):
        self.bno= ""
        self.phno= ""
        self.name= ""
        self.days= 0
        self.charge= 0
        
    def d_input(self,bno,phno,name,days):
        self.bno=bno
        self.phno=phno
        self.name=name
        self.days=days
        
    def compute(self):
        if self.days <= 5:
            self.charge = self.days * 500
        elif self.days <=10:
            self.charge = (5 * 500) + ((self.days - 5)* 400)
        else:
            self.charge = (5 * 500) + (5 * 400) + ((self.days - 10) * 200)
            
    def display(self):
        print("Bike no: ",self.bno)
        print("Phone no: ",self.phno)
        print("Name: ",self.name)
        print("Days: ",self.days)
        print("Charges: ",self.charge)

bno=int(input("Enter the bike no.: "))
phno=int(input("Enter the phone no.: "))
name=input("Enter the name: ")
days=int(input("Enter the days: "))

print("\n-----Details-----\n")

obj=Mobike()
obj.d_input(bno,phno,name,days)
obj.compute()
obj.display()

        
```

```python
#5.Create a class Book with attributes book id,book name,book price.Define appropriate constructor.
#Derive a class textbook from book with attribute publication and number of copies.
#Accept the details and display the details of the book.

class Book:
    def __init__(self,book_id,book_name,book_price):
        self.book_id=book_id
        self.book_name=book_name
        self.book_price=book_price
        
        
class TextBook(Book):
    def __init__(self,book_id,book_name,book_price,publication,copies):
        super().__init__(book_id,book_name,book_price)
        self.publication=publication
        self.copies=copies
            
    def display_details(self):
        print("Book Id: ",self.book_id)
        print("Book Name: ",self.book_name)
        print("Book Price: ",self.book_price)
        print("Publication: ",self.publication)
        print("Number of Copies: ",self.copies)
        
book_id=int(input("Enter the Book Id: "))
book_name=input("Enter the Book Name: ")
book_price=float(input("Enter the Book Price: "))
publication=input("Enter the publication: ")
copies=int(input("Enter the number of copies: "))
print("\n-----Book Details-----\n")

textbook=TextBook(book_id,book_name,book_price,publication,copies)
textbook.display_details()
        
```
